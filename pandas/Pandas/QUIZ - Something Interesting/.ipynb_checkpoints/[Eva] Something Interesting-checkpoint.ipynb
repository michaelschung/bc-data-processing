# Your "Thoughtful Queries"

```python
# Load your CSV file here, and preview the first 10-15 rows
netflix_titles = pd.read_csv('netflix_titles.csv', index_col = 2)

netflix_titles.head(15)
```

## Thoughtful Query #1

Question: How many movies made in the United States that are rated TV-MA were released between 2010 and 2019 (inclusive)?

```python
# Query 1 here
netflix_titles[(netflix_titles['country'] == 'United States') & (netflix_titles['rating'] == 'TV-MA') & (netflix_titles['release_year'] >= 2010) & (netflix_titles['release_year'] <= 2019)]
```

Insight: The table now only displays all the shows that were made in the United States with a rating of TV-MA and released between the years of 2010 and 2019. Although not all of the rows are displayed, at the bottom of the table, it's written "601 rows" which means there are 601 shows altogether that fit this description. 

## Thoughtful Query #2

Question: Which rating is most common for movies out of the following: TV-MA, TV-Y7, TV-Y, R, TV-14, G, TV-G, PG, TV-PG, PG-13, NC-17?

```python
# Query 2 here
tv_ma = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'TV-MA')])
tv_y7 = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'TV-Y7')])
tv_y = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'TV-Y')])
r = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'R')])
tv_14 = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'TV-14')])
g = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'G')])
tv_g = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'TV-G')])
pg = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'PG')])
tv_pg = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'TV-PG')])
pg_13 = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'PG-13')])
nc_17 = len(netflix_titles[(netflix_titles['type'] == 'Movie') & (netflix_titles['rating'] == 'NC-17')])

print(' TV-MA:', tv_ma, ' TV-Y7:', tv_y7, ' TV-Y:', tv_y, ' R:', r, ' TV-14:', tv_14, ' G:', g, ' TV-G:', tv_g, ' PG:', pg, ' TV-PG:', tv_pg, ' PG-13:', pg_13, ' NC-17:', nc_17)
```

Insight: The values at the bottom represent the frequency of each of the ratings of movies that show up in the data set. From these values, we can see that the TV-MA has the highest value of 1348 which means that it is the most common rating for movies on netflix.

## Thoughtful Query #3

Question: Were more movies released between 2000-2010 or 2011-2020? 

```python
# Query 3 here
req_1 = len(netflix_titles[(netflix_titles['release_year'] >= 2000) & (netflix_titles['release_year'] <= 2010) & (netflix_titles['type'] == "Movie")])
req_2 = len(netflix_titles[(netflix_titles['release_year'] >= 2011) & (netflix_titles['release_year'] <= 2020)& (netflix_titles['type'] == "Movie")])

print('# of movies released between:')
print('2000 & 2010:', req_1) 
print('2011 & 2020:',req_2)
```

Insight: From the data described above, you can see that between 2000 and 2010, a total of 623 movies were released and between 2011 and 2020, a total of 3305 were released. Based on this, the answer to the question is that more movies were released between 2011 and 2020.
