# Your "Thoughtful Queries"

```python
# Load your CSV file here, and preview the first 10-15 rows
pokemonMatchup = pd.read_csv("pokemon_type_matchup.csv", index_col = 0)
pokemonMatchup
```

## Thoughtful Query #1

Question: _(edit this cell to record your question)_

```python
# Query 1 here
#Which pokemon have the best(and worst) defensive type in the game? (least weaknesses and most resistances/immunities && most weaknesses and least resistances/immunities)
pokemon = pokemonMatchup.drop("Number", axis = 1)

cols = ['Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice', 'Fighting', 'Poison', 'Ground', 'Flying', 'Psychic', 'Bug', 'Rock', 'Ghost', 'Dragon', 'Dark', 'Steel', 'Fairy']#Can't select columns by index
for col in cols:
    pokemon[col] = pokemon[col].str.slice_replace(0,1, '')
    pokemon[col] = pokemon[col].astype(float)

defenseScore = pokemon.sum(axis=1)
pokemon['Defense'] = defenseScore

bestDefensivePokemondf = pokemon[pokemon["Defense"]==defenseScore.min()]
bestDefensivePokemon = bestDefensivePokemondf.index

worstDefensivePokemondf = pokemon[pokemon["Defense"]==defenseScore.max()]
worstDefensivePokemon = worstDefensivePokemondf.index

print('The best defensive pokemon are', end=" ")
for poke in bestDefensivePokemon:
    print(poke, end=", ")
print('and the weakest defensive pokemon are', end=" ")
for poke in worstDefensivePokemon:
    print(poke, end=", ")

#This code calculates a defense score for each pokemon(the lower the score the more resistances/immunities they have, adds that as a column to the original df and then uses that final common to search for pokemon with the highest and lowest defense values)
#The actual query finds the pokemon with highest and lowest defense scores
```

## Thoughtful Query #2

Question: _(edit this cell to record your question)_

```python
# Query 2 here
#What is the best(and worst) offensive type? (If a random pokemon is choosen, what type move is most likely to do the most damage?)
pokemon = pokemonMatchup.drop("Number", axis = 1)

cols = ['Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice', 'Fighting', 'Poison', 'Ground', 'Flying', 'Psychic', 'Bug', 'Rock', 'Ghost', 'Dragon', 'Dark', 'Steel', 'Fairy']#Can't select columns by index
for col in cols:
    pokemon[col] = pokemon[col].str.slice_replace(0,1, '')
    pokemon[col] = pokemon[col].astype(float)

offensiveType = pokemon.sum()
bestOffensiveType = offensiveType.idxmax()
worstOffensiveType = offensiveType.idxmin()

print('The best offensive type is', bestOffensiveType, 'and the worst offensive type is', worstOffensiveType)

#This code calculates the total offensive power of every type based on the pokemon in the game right now and finds the types with highest and lowest offensive power
#The actual query is finding the types with highest and lowest values
```

## Thoughtful Query #3

Question: _(edit this cell to record your question)_

```python
# Query 3 here
#What is the most common resistance and what percent of pokemon resist that type? (If a random pokemon is choosen, which type is it most likely to resist?)
pokemon = pokemonMatchup.drop("Number", axis = 1)

cols = ['Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice', 'Fighting', 'Poison', 'Ground', 'Flying', 'Psychic', 'Bug', 'Rock', 'Ghost', 'Dragon', 'Dark', 'Steel', 'Fairy']#Can't select columns by index
for col in cols:
    pokemon[col] = pokemon[col].str.slice_replace(0,1, '')
    pokemon[col] = pokemon[col].astype(float)

commonResistance = 0
for col in cols:
    resistances = len(pokemon[pokemon[col]<1])
    if resistances > commonResistance:
        commonResistance = resistances
        commonResistanceType = pokemon[col]

percentResist = round(commonResistance/len(commonResistanceType)*100, 2)
print(percentResist, '% of pokemon resist', commonResistanceType.name)

#This code calculates the most common resistance based on current pokemon and then the % of pokemon that have that restance
#The actual query finds the type with the current highest # of pokemon that resist it
```
