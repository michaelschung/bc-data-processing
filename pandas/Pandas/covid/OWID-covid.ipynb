```python
import pandas as pd
```

```python
data = pd.read_csv('vaccinations-by-manufacturer.csv', index_col=0)
data
```

```python
locations = data.groupby('location')
```

```python
us = locations.get_group('United States')
us
```

```python
us[us['vaccine'] == 'Pfizer/BioNTech']
```

```python
us['date'] = pd.to_datetime(us['date'])
us
```

```python
pfizer_mask = us['vaccine'] == 'Pfizer/BioNTech'
pre_march_mask = us['date'] < pd.to_datetime('2021-03-01')
pfizer = us[pfizer_mask & pre_march_mask]
pfizer
```
