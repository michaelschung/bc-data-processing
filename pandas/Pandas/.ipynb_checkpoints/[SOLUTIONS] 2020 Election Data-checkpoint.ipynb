## What percentage of the vote did Biden receive in Otsego County, NY?

Group the data by (state, county), calculate the percentage of the vote that each candidate received within each county, and add that data as a new column in the original DataFrame. Name this new column `pct_county`.

Then, query for Otsego County, NY, to see what percentage Biden received there.

```python
# Group the data by (state, country)
counties = election.groupby(['state', 'county'])
counties.groups
```

```python
# Write a method that calculates the percentage of each value in a group with respect to the whole group.
# This is very similar to the pct function from Investigating Transform.
def pct(x):
    return x / sum(x)
```

```python
# Use the above method to calculate the percentage of each value with respect to its group, and add
# this information as a new column in the original DataFrame.
election['pct_vote'] = counties['total_votes'].transform(pct)
new_york = election.loc['New York']
new_york[new_york['county'] == 'Otsego County']
```

## Would Trump have won if he received the Libertarian votes?

A common sentiment expressed in the aftermath of the 2020 election was that Trump would have won if those who voted Libertarian had voted for him instead. Use the data to determine if that's a true statement.

*Hint: We already know that if the few key swing states (Arizona, Georgia, Pennsylvania, Wisconsin) had flipped, then Trump would have won the electoral college vote. Thus, the easiest way to solve this problem is to check to see if those particular states would have flipped.*

```python
# Group the data by state
states = election.groupby('state')
states.groups
```

```python
# Find each state that would have flipped if the Libertarian votes had gone Republican
# Loop through each state
for name, state in states:
    # Group state by candidate
    state_cand = state.groupby('candidate')
    # Find total vote count per candidate
    totals = state_cand['total_votes'].sum()
    # Grab each candidate's total
    biden = totals['Joe Biden']
    trump = totals['Donald Trump']
    jorg = totals['Jo Jorgensen']
    # Calculate difference between biden and trump
    diff = biden - trump
    # Only print if Jorgensen's count is greater than Biden's lead
    if (biden > trump and jorg > diff):
        hypothetical_trump_lead = trump + jorg - biden
        print(name, hypothetical_trump_lead)
```

Conclusion: Yes. If all Libertarian voters had voted for Trump, then he would have won the popular vote for these four states, and thus their electoral votes.
