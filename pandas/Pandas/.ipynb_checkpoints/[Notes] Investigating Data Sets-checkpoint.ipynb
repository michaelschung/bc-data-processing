```python
# Read flights_small.csv into a DataFrame called flights
flights = pd.read_csv('flights_small.csv')
flights.head()
```

```python
a_count = len(flights[flights['CANCELLATION_REASON'] == 'A'])
b_count = len(flights[flights['CANCELLATION_REASON'] == 'B'])
c_count = len(flights[flights['CANCELLATION_REASON'] == 'C'])
d_count = len(flights[flights['CANCELLATION_REASON'] == 'D'])

print(a_count, b_count, c_count, d_count)
```
