```python
import pandas as pd
import leather
```

# Election Data from Kent County, MI

```python
# Import data
straight_party = pd.read_csv('kent-county-straight-party.csv', index_col=0)
individual = pd.read_csv('kent-county-individual.csv', index_col=0)
straight_party
```

## Part 1: Recreate Dr. Ayyadurai's graph

In the video, Dr. Ayyadurai presents the graph below to argue that Trump's votes in Kent County, MI were being transferred to Biden. As our first task, let's recreate this graph from the data that we have.



```python
# Recreate Dr. Ayyadurai's graph
def get_pct_rep(row):
    dems = row['Democratic Party']
    reps = row['Republican Party']
    pct_rep = reps / (dems + reps) * 100
    return pct_rep

straight_party.drop('TOTAL', inplace=True)
pct_rep = straight_party.apply(get_pct_rep, axis=1)
```

```python
def get_pct_trump(row):
    biden = row['Joseph R. Biden w/ Kamala D. Harris']
    trump = row['Donald J. Trump w/ Michael R. Pence']
    pct_trump = trump / (biden + trump) * 100
    return pct_trump

individual.drop('TOTAL', inplace=True)
pct_trump = individual.apply(get_pct_trump, axis=1)
```

```python
# Create list of tuples as data points
points = list(zip(pct_rep, pct_trump - pct_rep))
# Create scatter plot using data points
chart = leather.Chart('Dr. Ayyadurai\'s graph')
chart.add_dots(points, radius=3, fill_color='red')
chart.to_svg()
```
