```python
shark = input("Do you like sharks? ")
def likeSharks(shark):
    if shark == 'no':
        while shark != "yes":
            shark = input("Do you like sharks now? ")
            if shark != 'yes':
                print(shark)
        for i in range(100):
            print('Then why hath you denied it, blasphemer?')
    elif shark == 'yes':
        print('As you should young warrior')
        return
    else:
        print("it's a yes or no question... blasphemer")

likeSharks(shark)
```
