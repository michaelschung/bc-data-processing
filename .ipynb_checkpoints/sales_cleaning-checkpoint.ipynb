```python
import pandas as pd
```

```python
order_leads = pd.read_csv(
    'https://raw.githubusercontent.com/michaelschung/bc-data-processing/master/datasets/sales/order_leads.csv',
    parse_dates = [3]
)
sales_team = pd.read_csv(
    'https://raw.githubusercontent.com/michaelschung/bc-data-processing/master/datasets/sales/sales_team.csv',
    parse_dates = [3]
)
df = pd.merge(
  order_leads,
  sales_team,
  on=['Company Id','Company Name']
)
df = df.rename(
  columns={'Order Value':'Val','Converted':'Sale'}
)
```

```python
df.head(20)
```

```python
df.to_csv('./datasets/sales/sales.csv', index_label=False)
```
