```python
import pandas as pd
```

```python
olympics_url = 'https://raw.githubusercontent.com/michaelschung/bc-data-processing/master/datasets/olympics.csv'

olymp = pd.read_csv(olympics_url, index_col=0, skiprows=1)
olymp.head()
```

```python
columns = olymp.columns
columns
```

```python
for col in columns:
  # If the column name starts with '01'...
  if col[:2]=='01':
    # ...rename the column `Gold`, followed by the slice `.1` or `.2`
    # inplace=True forces the change to apply to the original DataFrame
    olymp.rename(columns={col:'Gold' + col[4:]}, inplace=True)
  elif col[:2]=='02':
    olymp.rename(columns={col:'Silver' + col[4:]}, inplace=True)
  elif col[:2]=='03':
    olymp.rename(columns={col:'Bronze' + col[4:]}, inplace=True)

  # For good measure, let's replace the numero sign with a normal # sign.
  if col[:1]=='№':
    olymp.rename(columns={col:'#' + col[1:]}, inplace=True) 

olymp
```

```python
olymp.to_csv('./datasets/olympics-clean.csv')
```
