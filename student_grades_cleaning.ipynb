```python
import pandas as pd
import datetime as dt
```

```python
grades = pd.read_csv('datasets/grades.csv', index_col=0)
grades
```

```python
new_col_names = [name[:6] + name[10:] for name in grades.columns]
new_col_names
```

```python
grades.set_axis(new_col_names, axis=1, inplace=True)
grades
```

```python
sub_cols = [name for name in grades.columns if name.find('sub') >= 0]
sub_cols
```

```python
assign1_sub_col = grades[sub_cols[0]]
assign1_sub_col[0]
```

```python
assign1_sub_col = grades[sub_cols[0]]
dt.datetime.strptime(assign1_sub_col[0][:-3], '%Y-%m-%d %H:%M:%S.%f')
```

```python
for sub_col in sub_cols:
    sub_times_raw = grades[sub_col]
    datetimes = [dt.datetime.strptime(time[:-3], '%Y-%m-%d %H:%M:%S.%f') for time in sub_times_raw]
    years = [x.year for x in datetimes]
    months = [x.month for x in datetimes]
    days = [x.day for x in datetimes]
    hours = [x.hour for x in datetimes]
    minutes = [x.minute for x in datetimes]
    seconds = [x.second for x in datetimes]
    
    col_name_stub = sub_col[:11] + '_'
    pieces = ['year', 'month', 'day', 'hour', 'min', 'sec']
    col_names = [col_name_stub + piece for piece in pieces]
    final_columns = [years, months, days, hours, minutes, seconds]
    
    for i, col_name in enumerate(col_names):
        grades[col_name] = final_columns[i]

grades
```

```python
grades.drop(sub_cols, axis=1, inplace=True)
grades
```

```python
sorted_col_names = []
for i in range(1, 7):
    sorted_col_names.append(f'assign{i}_grade')
    for piece in ['year', 'month', 'day', 'hour', 'min', 'sec']:
        sorted_col_names.append(f'assign{i}_sub_{piece}')
sorted_col_names
```

```python
grades.reindex(columns=sorted_col_names, inplace=True)
grades
```
